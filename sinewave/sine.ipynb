```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generate_uint16_sine_table(num_samples: int = 1024, amplitude: int = 32767, offset:int = 32767) -> tuple[np.ndarray]:
    num_values = num_samples
    x = np.linspace(0, 2 * np.pi, num_values, endpoint=False)
    y = np.round(amplitude * np.sin(x) + offset).astype(int)
    return x,y
```

```python
num_samples: int = 1024
amplitude = 32767
offset = 32767
x,y = generate_uint16_sine_table(num_samples, amplitude,offset)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y, label="Sine wave LUT values", color="blue")
plt.title("Generated Sine Wave LUT")
plt.xlabel("Angle (radians)")
plt.ylabel("Amplitude (16-bit)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
row_size = 25
with open("sinewave_values.txt","w+") as file:
    for i in range(0, len(y.tolist()), row_size):
        row = y.tolist()[i:i + row_size]
        file.write(",".join(map(str, row)) + "\n")
```
